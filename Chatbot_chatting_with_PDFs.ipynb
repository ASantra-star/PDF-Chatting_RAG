```python
from langchain.chains import RetrievalQA
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import LlamaCpp
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.document_loaders import PyPDFDirectoryLoader
```

```python
#load pdf files
loader = PyPDFDirectoryLoader("/content/drive/MyDrive/data/data2/")
data = loader.load()
```

```python
print(data)
```

```python
#Step 05: Split the Extracted Data into Text Chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=20)

text_chunks = text_splitter.split_documents(data)
```

```python
#Step 06:Downlaod the Embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
#Step 08: Create Embeddings for each of the Text Chunk
vector_store = FAISS.from_documents(text_chunks, embedding=embeddings)
```

```python
#Import Model
llm = LlamaCpp(
    streaming = True,
    model_path="/content/drive/MyDrive/data/mistral-7b-instruct-v0.1.Q4_K_M.gguf",
    temperature=0.75,
    top_p=1,
    verbose=True,
    n_ctx=4096
)
```

```python
# Chatbot loop
def chatbot():
    print("Hello! I'm your chatbot. Ask me anything about the Language of Flowers.")
    while True:
        query = input("You: ")
        if query.lower() in ['exit', 'quit', 'bye']:
            print("Goodbye!")
            break
        try:
            response = qa.run(query)
            print(f"Bot: {response}")
        except Exception as e:
            print(f"An error occurred: {e}")

# Start the chatbot
chatbot()
```

```python
import time

def measure_time(chunk_size, chunk_overlap):
    # Split text
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    text_chunks = text_splitter.split_documents(data)

    # Create embeddings and vector store
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vector_store = FAISS.from_documents(text_chunks, embedding=embeddings)

    # Import Model
    llm = LlamaCpp(
        streaming=True,
        model_path="/content/drive/MyDrive/data/mistral-7b-instruct-v0.1.Q4_K_M.gguf",
        temperature=0.75,
        top_p=1,
        verbose=True,
        n_ctx=4096
    )

    # Create RetrievalQA
    qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_store.as_retriever(search_kwargs={"k": 2}))

    # Measure time for a sample query
    query = "What is the symbolism of honeysuckle?"
    start_time = time.time()
    response = qa.run(query)
    end_time = time.time()

    print(f"Chunk size: {chunk_size}, Chunk overlap: {chunk_overlap}")
    print(f"Time taken: {end_time - start_time:.2f} seconds")
    print(f"Response: {response}")

# Test different configurations
print("Testing configuration 1:")
measure_time(chunk_size=500, chunk_overlap=10)

print("\nTesting configuration 2:")
measure_time(chunk_size=1000, chunk_overlap=20)

print("\nTesting configuration 3:")
measure_time(chunk_size=2000, chunk_overlap=50)

print("Testing configuration 4:")
measure_time(chunk_size=500, chunk_overlap=20)

print("\nTesting configuration 5:")
measure_time(chunk_size=1000, chunk_overlap=10)

print("\nTesting configuration 6:")
measure_time(chunk_size=2000, chunk_overlap=20)
```

```python
import time

def measure_time(chunk_size, chunk_overlap):
    # Split text
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    text_chunks = text_splitter.split_documents(data)
    num_chunks = len(text_chunks)  # Count the number of chunks

    # Create embeddings and vector store
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vector_store = FAISS.from_documents(text_chunks, embedding=embeddings)

    # Import Model
    llm = LlamaCpp(
        streaming=True,
        model_path="/content/drive/MyDrive/data/mistral-7b-instruct-v0.1.Q4_K_M.gguf",
        temperature=0.75,
        top_p=1,
        verbose=True,
        n_ctx=4096
    )

    # Create RetrievalQA
    qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_store.as_retriever(search_kwargs={"k": 2}))

    # Measure time for a sample query
    query = "Describe some stories related to White Rose"
    start_time = time.time()
    response = qa.run(query)
    end_time = time.time()

    print(f"Chunk size: {chunk_size}, Chunk overlap: {chunk_overlap}")
    print(f"Number of chunks: {num_chunks}")
    print(f"Time taken: {end_time - start_time:.2f} seconds")
    print(f"Response: {response}")

# Test different configurations
print("Testing configuration 1:")
measure_time(chunk_size=500, chunk_overlap=10)

print("\nTesting configuration 2:")
measure_time(chunk_size=1000, chunk_overlap=20)

print("\nTesting configuration 3:")
measure_time(chunk_size=2000, chunk_overlap=50)

print("\nTesting configuration 4:")
measure_time(chunk_size=500, chunk_overlap=20)

print("\nTesting configuration 5:")
measure_time(chunk_size=1000, chunk_overlap=10)

print("\nTesting configuration 6:")
measure_time(chunk_size=2000, chunk_overlap=20)

print("\nTesting configuration 7:")
measure_time(chunk_size=10000, chunk_overlap=10)

print("\nTesting configuration 8:")
measure_time(chunk_size=10000, chunk_overlap=50)
```
